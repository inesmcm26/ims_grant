```python
f1_to_plot
```

```python
import plotly.express as px
import ast
import pandas as pd

# Assuming you have a DataFrame named 'df' with algorithm names as columns and F1 scores as lists in each cell

# Extracting the F1 scores for train and validation sets
train_scores = f1_to_plot.loc['Train F1']
validation_scores = f1_to_plot.loc['Val F1']

# Converting the strings to lists
train_scores = train_scores.apply(ast.literal_eval)
validation_scores = validation_scores.apply(ast.literal_eval)

# Creating a new DataFrame in the required format
data_train = pd.DataFrame()
data_val = pd.DataFrame()

for col in train_scores.index:
    data_train = pd.concat([data_train, pd.DataFrame({'Algorithm': [col] * len(train_scores[col]),
                                                      'Score': train_scores[col],
                                                      'Set': 'Train'})], ignore_index=True)
    data_val = pd.concat([data_val, pd.DataFrame({'Algorithm': [col] * len(validation_scores[col]),
                                                  'Score': validation_scores[col],
                                                  'Set': 'Validation'})], ignore_index=True)

data = pd.concat([data_train, data_val])

# Defining custom colors for the box plot
color_map = {'Train': 'rgb(64, 152, 191)', 'Validation': 'rgb(108, 191, 64)'}

# Creating the box plot using Plotly Express
fig = px.box(data, x='Algorithm', y='Score', color='Set', color_discrete_map=color_map)

# Updating the layout
fig.update_layout(title='F1 Measure Comparison',
                  xaxis_title='Algorithm',
                  yaxis_title='F1 Measure',
                  legend_title='Set',
                  plot_bgcolor='rgb(240, 240, 240)',
                  height=600)  # Adjust the height value as per your preference

# Showing the plot
fig.show()
```

## Statistical significance comparison of results

The Willcoxon signed-rank test will be used to asses the statistical significance of the performance of the best models of each algorithm.

The null hypothesis of the test is that the two models have equal performance on the classification task, and the alternative hypothesis is that they have different performance. If the p-value is lower than a threshold (0.05, in our case), then it is extremelly unlikely that the null hypothesis is true. Therefore, for every p-value < 0.05, the difference in performance of the two models is significant.

```python
# convert string to list
best_scores['Train F1'] = best_scores['Train F1'].apply(eval)
```

```python
best_scores['Val F1'] = best_scores['Val F1'].apply(eval)
```

```python
# create dataframe to store the p-values
p_values_df = pd.DataFrame(columns=best_models, index=best_models)

p_values_df
```

```python
for i, model1 in enumerate(best_models):
    for model2 in best_models[i:]:
        # skip if the models are the same
        if model1 == model2:
            continue
        # calculate the p-value
        statistic, p_value = wilcoxon(best_scores.loc[model1, 'Val F1'], best_scores.loc[model2, 'Val F1'])

        # store the p-value in the dataframe
        p_values_df.loc[model1, model2] = p_value

p_values_df.style.applymap(lambda x: 'background-color : darkgreen' if x < 0.05 else 'background-color : None')
```

- MLP performance not significant comparing with DT and GB
